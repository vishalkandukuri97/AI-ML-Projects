```python
import cv2
import matplotlib.pyplot as plt
image=cv2.imread('/content/1.png')
plt.imshow(image)
```

```python
grey_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image, cmap='gray')
```

```python
sift = cv2.SIFT_create()
keypoints, descriptors=sift.detectAndCompute(grey_image, None)
```

```python
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS, color=(0, 255, 0))
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.imshow(cv2.cvtColor(image_with_keypoints, cv2.COLOR_BGR2RGB))
```

```python
plt.title("SIFT Feature Keypoints")

```

```python
plt.axis('off')
```

```python
plt.show()
```
