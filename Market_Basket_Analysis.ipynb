```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install --quiet mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = [
    ["milk","bread","butter"],
    ["milk","bread"],
    ["milk","bread","eggs"],
    ["bread","butter"],
    ["milk","eggs"],
    ["bread","eggs","butter"],
    ["milk","bread","butter","eggs"],
    ["bread","butter"],
    ["milk","bread","eggs"],
    ["bread","butter"]
]
```

```python
df=pd.DataFrame(data)
```

```python
basket = df.stack().str.get_dummies().groupby(level=0).sum()
ITEMS = sorted(list(basket.columns))
```

```python
def analyze(item, min_support, min_conf, top_k):
    freq = apriori(basket, min_support=min_support, use_colnames=True)
    if freq.empty: return "⚠️ No frequent itemsets found.", pd.DataFrame()

    rules = association_rules(freq, metric="confidence", min_threshold=min_conf)
    rules = rules.sort_values(by=["confidence","lift"], ascending=[False, False])
    mask = rules["antecedents"].apply(lambda s: item in s)
    recs = rules[mask]
    if recs.empty: return f"⚠️ No rules found for {item}.", pd.DataFrame()

    recs = recs.head(top_k)
    def fs_to_str(fs): return ", ".join(sorted(list(fs)))
    show = pd.DataFrame({
        "Antecedents": recs["antecedents"].apply(fs_to_str),
        "Consequents": recs["consequents"].apply(fs_to_str),
        "Support": recs["support"].round(2),
        "Confidence": recs["confidence"].round(2),
        "Lift": recs["lift"].round(2)
    })
    explanation = "\n".join([
        f"- If basket has *{row['Antecedents']}, it also has *{row['Consequents']}** "
        f"(conf: {row['Confidence']:.2f}, lift: {row['Lift']:.2f})"
        for _, row in show.iterrows()
    ])
    return explanation, show
```

```python
import gradio as gr
```

```python
with gr.Blocks() as demo:
    gr.Markdown("## 🏃‍♂️ Mini Retail Basket Analysis (Apriori - Fast Demo)")
    with gr.Row():
        item_dd = gr.Dropdown(choices=ITEMS, value=ITEMS[0], label="Select Item")
    with gr.Row():
        min_support = gr.Slider(0.1, 0.9, value=0.3, step=0.05, label="Min Support")
        min_conf = gr.Slider(0.2, 1.0, value=0.5, step=0.05, label="Min Confidence")
        top_k = gr.Slider(1, 10, value=5, step=1, label="Top K Rules")
    run_btn = gr.Button("Run Apriori", variant="primary")
    out_text = gr.Markdown()
    out_df = gr.Dataframe()
    run_btn.click(analyze, [item_dd, min_support, min_conf, top_k], [out_text, out_df])
```

```python
demo.launch()
```
